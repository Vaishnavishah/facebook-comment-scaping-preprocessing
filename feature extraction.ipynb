```python
A=[" Peace to all politics"," Pray for the world ", "!!!  Way to go President Trump.  Keep those Dems guessing how you do such a great job.  They have not done much of anything in almost 4 years.  Sad","Great job ! Now if only those God forsaken protesters ... looters .. and rioters ... would actually get a real job the numbers would look even better !"," Many make promises, few deliver.","...and they said jobs wouldn't come back until the Fall. They forget who's in charge.","Democrats are scrambling trying to create something negative about jobs report."]
```

```python
import pandas as pd
data = pd.DataFrame({'comment_text':A})
```

```python
data['comment_text']
```

```python
#Number of Words
data['word_count'] = data['comment_text'].apply(lambda x: len(str(x).split(" ")))
data[['comment_text','word_count']].head()
```

```python
#Number of characters
data['char_count'] = data['comment_text'].str.len() ## this also includes spaces
data[['comment_text','char_count']].head()
```

```python
#Average Word Length
#Number of characters(without space count)/Total number of words
def avg_word(sentence):
  words = sentence.split()
  print(words)
  print(len(words))
  print(sum(len(word) for word in words))
  return (sum(len(word) for word in words)/len(words))

data['avg_word'] = data['comment_text'].apply(lambda x: avg_word(x))
data[['comment_text','avg_word']].head()
```

```python
#Number of stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')

data['stopwords'] = data['comment_text'].apply(lambda x: len([x for x in x.split() if x in stop]))
data[['comment_text','stopwords']].head()
```

```python
#Number of special characters
data['hastags'] = data['comment_text'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
data[['comment_text','hastags']].head()
```

```python
#Number of numerics
data['numerics'] = data['comment_text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
data[['comment_text','numerics']].head()
```

```python
#Number of Uppercase words
data['upper'] = data['comment_text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
data[['comment_text','upper']].head()
```

```python
pos_family = {
    'noun' : ['NN','NNS','NNP','NNPS'],
    'pron' : ['PRP','PRP$','WP','WP$'],
    'verb' : ['VB','VBD','VBG','VBN','VBP','VBZ'],
    'adj' :  ['JJ','JJR','JJS'],
    'adv' : ['RB','RBR','RBS','WRB']
}

# function to check and get the part of speech tag count of a words in a given sentence
from textblob import TextBlob, Word, Blobber
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
def check_pos_tag(x, flag):
    cnt = 0
    try:
        wiki = TextBlob(x)
        for tup in wiki.tags:
            ppo = list(tup)[1]
            if ppo in pos_family[flag]:
                cnt += 1
                print(ppo, tup)
    except:
        pass
    return cnt

data['noun_count'] = data['comment_text'].apply(lambda x: check_pos_tag(x, 'noun'))
data['verb_count'] = data['comment_text'].apply(lambda x: check_pos_tag(x, 'verb'))
data['adj_count'] = data['comment_text'].apply(lambda x: check_pos_tag(x, 'adj'))
data['adv_count'] = data['comment_text'].apply(lambda x: check_pos_tag(x, 'adv'))
data['pron_count'] = data['comment_text'].apply(lambda x: check_pos_tag(x, 'pron'))
data[['comment_text','noun_count','verb_count','adj_count', 'adv_count', 'pron_count' ]].head()
```
